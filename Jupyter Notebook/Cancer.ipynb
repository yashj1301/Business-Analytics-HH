```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cancer_data=load_breast_cancer()
cancer_data.feature_names
```

```python
cancer_data.target #target dataset inside cancer_data is stored in the form of an array
```

```python
cancer = pd.DataFrame(data=cancer_data.data, columns=cancer_data.feature_names)
cancer.head(10) #converted to pandas dataframe
```

```python
cancer['Target']=cancer_data.target #creating a new column
cancer['Target'].value_counts()
```

```python
columns=cancer.columns.tolist()
columns
```

```python
lr=LogisticRegression(class_weight='balanced') #putting logistic regression in the analysis
```

```python
from sklearn.model_selection import train_test_split
trainX,testX,trainY,testY=train_test_split(cancer_data.data,cancer_data.target)
lr.fit(trainX,trainY)
```

```python
y_pred=lr.predict(testX) #predicting values
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_pred, testY) #using confusion matrix for validation
```

```python
accuracy_score(y_pred,testY).round(3) #accuracy score from confusion matrix 
```
